```python
corr_matrix = df.drop(columns=['user_id']).corr()

corr_matrix['Final_Exam'].sort_values(ascending=False)
```

Based on the correlation values, I would choose A4 and A7 to predict the final exam grade. These two assignments have the strongest correlation with Final_Exam, which suggests they are the best predictors among the assignments

```python
df_filled = df.fillna(df.mean(numeric_only=True))

df_filled.isnull().sum()
```

I used mean imputation to handle missing values. After replacing missing values with the column averages, there were no missing values left in the dataset.

```python
df.describe()
```

I looked at the maximum values for each assignment and found some grades that were much higher than normal. For example, A4 has a maximum grade of 188.2 and A6 has a maximum grade of 183.6. These could be outliers. Since the lab mentions that some assignments had bonus marks, I decided to keep these values instead of removing them.

```python
df_median = df.fillna(df.median(numeric_only=True))

df_median.isnull().sum()
```

I explored the dataset, checked for missing values, and looked at the correlations between assignments and the final exam. A4 and A7 had the strongest relationship with the final exam, so I chose them as the best predictors. I used mean imputation to fill missing values because it is simple and allowed me to keep all students in the dataset. I also tested median imputation. It removed all missing values and produced similar results to mean imputation. I decided to use mean imputation because it is simple and worked well for this dataset.
