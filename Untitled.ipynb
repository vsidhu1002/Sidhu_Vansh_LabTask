# Part 5: Arithmetic and Data Alignment

This code creates a new column by multiplying the Fare column by 0.92 to convert the values into Euros.

```python
df["Fare_Euros"] = df["Fare"] * 0.92

df[["Fare", "Fare_Euros"]]
```

```python
df["Age_Difference_From_Average"] = df["Age"] - df["Age"].mean()

df[["Age", "Age_Difference_From_Average"]]
```

```python
selected_columns = df[["Age", "Fare"]]

custom_array = np.array([10, 20])

selected_columns - custom_array
```

# Part 6: Applying Functions

This code uses apply() to count how many characters are in each passenger's name.

```python
# Create a new column with number of characters in each name
df["Name_Length"] = df["Name"].apply(len)

df[["Name", "Name_Length"]]
```

```python
# Calculate difference between fare and age
df["Fare_Minus_Age"] = df["Fare"] - df["Age"]

df[["Fare", "Age", "Fare_Minus_Age"]]
```

```python
df.round(1)
```

# Part 7: Mapping and Transformation

```python
df["Sex_Code"] = df["Sex"].map({"male": 0, "female": 1})

df[["Sex", "Sex_Code"]]
```

```python

def categorize_age(age):
    if pd.isna(age):
        return "Unknown"
    elif age < 18:
        return "Child"
    elif age < 60:
        return "Adult"
    else:
        return "Senior"

df["Age_Group"] = df["Age"].apply(categorize_age)

df[["Age", "Age_Group"]]
```

This function checks each age value and places it into a group such as Child, Adult, Senior, or Unknown.

# Part 8: Pure Python Implementation and Time Comparison

```python
import csv

data = []

with open("titanic.csv", newline='', encoding='utf-8') as file:
    reader = csv.DictReader(file)

    for row in reader:
        data.append(row)

print(data[:2])
```

```python
names = []

for row in data:
    names.append(row["Name"])

print(names[:10])
```
